### Handle Missing Values

#### 1 There are different ways to handle misssing values:
    1. Mean
    2. Median
    3. Mode

```python
df['price'].sum() / 7
```

### 1. Mean()

```python
df1 = df['price'].mean()
```

```python
df1
```

```python
df['price'].fillna(df1, inplace=True)
print(df.to_string())
```

### 3. Mode()
