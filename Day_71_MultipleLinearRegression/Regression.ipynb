# Deal with Missing Data

1. Drop Data:
    1. drop the whole row
    2. drop the whole column
2. replace data:
    1. replace by mean
    2. replace by mode [frequency]
    3. replace by median
    4. replace it based on the other function

1. Normalize losses - 41
2. stroke
3. bore
4. horsepower
5. peak-rpm
6. price
7. num-of-doors

```python
avg_norm_loss = df['normalized-losses'].astype('float').mean(axis=0)
print("Average value of Normalaized losses: ", avg_norm_loss)
```

```python
df['normalized-losses'].replace(np.nan, avg_norm_loss, inplace= True)
```

```python
avg_bore = df['bore'].astype('float').mean(axis=0)
print("Average value of bore values: ", avg_bore)
```

```python
df['bore'].replace(np.nan, avg_bore, inplace= True)
```

```python
avg_stroke = df['stroke'].astype('float').mean(axis=0)
print("Average value of stroke values: ", avg_stroke)
```

```python
df['stroke'].replace(np.nan, avg_stroke, inplace= True)
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average value of horsepower values: ", avg_horsepower)
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace= True)
```

```python
avg_peakrpm = df['peak-rpm'].astype('float').mean(axis=0)
print("Average value of peak-rpm values: ", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_stroke, inplace= True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

### Replace missing value by frequent value [mode]

```python
df['num-of-doors'].replace(np.nan, 'four',inplace = True)
```

```python
df.head()
```

```python
df.dropna(subset=['price'], axis = 0, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

### Lets see the data types for each column

```python
df.dtypes
```

```python
df['bore'] = df['bore'].astype('float')
df['stroke'] = df['stroke'].astype('float')
df['normalized-losses'] = df['normalized-losses'].astype('int64')
df['peak-rpm'] = df['peak-rpm'].astype('float')
df['price'] = df['price'].astype('float')
```
