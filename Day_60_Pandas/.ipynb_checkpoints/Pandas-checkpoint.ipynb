### Replace Using Mead, Mode, Median

### 1. Mean():
    1. The average value (sum of all values divide by number of rows)

```python
df = pd.read_csv('car_sales_data.csv')
```

```python
df.head()
```

```python
x = df['price'].mean()
df['price'].fillna(x, inplace=True)
```

```python
df.head()
```

### 2. Median():
    1. In the specified feature the middle values select as median value ,first you have sorted values in ascending order 

```python
df = pd.read_csv('car_sales_data.csv')
```

```python
df.head()
```

```python
x = df['price'].median()
df['price'].fillna(x, inplace=True)
```

```python
df.head()
```

### 3. Mode():
    1. The values appers most frequently

```python
df =pd.read_csv('car_sales_data.csv')
```

```python
df.head()
```

```python
x = df['price'].mode()[0]
df['price'].fillna(x, inplace=True)
```

```python
df.head()
```

```python
df1 = df.copy()
```
