```python
df = df.drop(['normalized-losses','num-of-doors','body-style','aspiration','num-of-doors','make','fuel-type','drive-wheels','engine-location','engine-type'],axis=1)
```

```python
df[['num-of-cylinders']]
```

```python
df['num-of-cylinders'].unique()
```

```python
df['num-of-cylinders'] = df['num-of-cylinders'].replace({'two':2,'three':3,'four':4,'five':5,'six':6,'eight':8, 'twelve':12})
```

```python
df['num-of-cylinders'].uniqueque()
```

### Encoding:
    1. Object to Numerical
    2. Label Encoding
    3. Numerical or categorical
    4. One Hot Encoding
    5. Binary Encoding
    6. Hash Encoding

```python
df.T
```

```python
df = df.replace('?',np.nan)
```

```python
df.head(20)
```

```python
df['bore'] = df['bore'].astype('float64')
df['stroke'] = df['stroke'].astype('float64')
df['horsepower'] = df['horsepower'].astype('float64')
df['peak-rpm'] = df['peak-rpm'].astype('float64')
df['price'] = df['price'].astype('float64')
```

```python
df.dtypes
```

```python
df.head(50)
```

```python
df['bore'] = df['bore'].fillna(df['bore'].median())
df['stroke'] = df['stroke'].fillna(df['stroke'].median())
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
df['peak-rpm'] = df['peak-rpm'].fillna(df['peak-rpm'].median())
df['price'] = df['price'].fillna(df['price'].median())
```

```python
df.describe().T
```

```python
sns.pairplot(df, diag_kind='kde')
```

```python
df.shape
```

```python
x = df.drop('price', axis=1)
x
```

```python
y = df[['price']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size =0.2)
```

```python
x_test.shape
```

```python
# Training >> 164
# Testing >> 41
```

```python
linear_regression = LinearRegression()
```

```python
linear_regression.fit(x_train,y_train)
```

```python
y_pred = linear_regression.predict(x_test)
```
