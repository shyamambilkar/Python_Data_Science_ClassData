## 3. Poission Distribution

1. Poission Distributin is Descrete Data Distribution
2. how many times event can be happen in specified time:
    1. Example: person can be eat in day thrice or fifth time what is the probability
    
3. It has two paramters:
    1. lam - rate or number of occurances
    2. size - shape of returned array

```python
from numpy import random

my_num = random.poisson(lam = 2, size = 10)
my_num
```

```python
sns.distplot(random.poisson(lam = 2, size = 1000), kde=False)
plt.show()
```
