```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_array = np.array([10,20,30,40,50,60])

plt.pie(my_array)
plt.show()
```

```python
my_array = np.array([10,20,30,40,50,60])
my_labels = ['Apple', 'Cherry','Watermelon', 'Mango', 'Pineapple','Banana']

plt.pie(my_array, labels= my_labels)
plt.show()
```

```python
my_array = np.array([10,20,30,40,50,60])
my_labels = ['Apple', 'Cherry','Watermelon', 'Mango', 'Pineapple','Banana']

plt.pie(my_array, labels= my_labels, startangle= 0)
plt.show()
```

```python
my_array = np.array([10,20,30,40,50,60])
my_labels = ['Apple', 'Cherry','Watermelon', 'Mango', 'Pineapple','Banana']
my_explode = [0,0,0,0,0.1,0.2]

plt.pie(my_array, labels= my_labels, startangle= 0, explode = my_explode)
plt.show()
```

```python
my_array = np.array([10,20,30,40,50,60])
my_labels = ['Apple', 'Cherry','Watermelon', 'Mango', 'Pineapple','Banana']
my_explode = [0,0,0,0,0.1,0.2]

plt.pie(my_array, labels= my_labels, startangle= 0, explode = my_explode, shadow=True)
plt.show()
```

```python
my_array = np.array([10,20,30,40,50,60])
my_labels = ['Apple', 'Cherry','Watermelon', 'Mango', 'Pineapple','Banana']
my_explode = [0,0,0,0,0.1,0.2]
my_colors = ['black', 'yellow','blue','orange','purple','#4CAF50']

plt.pie(my_array, labels= my_labels, startangle= 0, explode = my_explode, shadow=True, colors = my_colors)
plt.legend(title = "Fruit Sales")
plt.show()
```
