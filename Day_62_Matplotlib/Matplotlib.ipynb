### colormap [cmap]

```python
x = np.array([5,6,8,10,80,45,63,78,95,62,35,98,105])
y = np.array([45,47,36,52,100,105,110,115,120,125,130,135,140])
color = np.array([0,10,20,30,40,45,50,50,55,60,65,70,100])

plt.scatter(x,y,c=color,cmap='Oranges')
plt.colorbar()
plt.show()
```

```python
x = np.array([5,6,8,10,80,45,63,78,95,62,35,98,105])
y = np.array([45,47,36,52,100,105,110,115,120,125,130,135,140])
color = np.array([0,10,20,30,40,45,50,50,55,60,65,70,100])
sizes = np.array([45,47,36,52,100,105,110,115,600,500,130,135,140])

plt.scatter(x,y,c=color,cmap='Oranges', s=sizes)
plt.colorbar()
plt.show()
```

```python
x = np.array([5,6,8,10,80,45,63,78,95,62,35,98,105])
y = np.array([45,47,36,52,100,105,110,115,120,125,130,135,140])
color = np.array([0,10,20,30,40,45,50,50,55,60,65,70,100])
sizes = np.array([45,47,36,52,100,105,110,115,600,500,130,135,140])

plt.scatter(x,y,c=color,cmap='Oranges', s=sizes, alpha=0.5)
plt.colorbar()
plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes =  5 * np.random.randint(100, size=(100))

plt.scatter(x,y, c=colors,s= sizes, alpha=0.6, cmap='RdYlGn_r')
```
