# Python Loop:

There are different types of Loops:
    1. For Loop
    2. While Loop

### 1. For Loop

1. A for Loop is used to itrating elements sequencially such as (List, Tuples, Dictinary, Set or String)
2. For loop uses for keyword
3. set of elements execute set of statement

```python
my_list = ['apple','mango','banana','cherry', 'orange','pineapple']

for x in my_list:
    print(x)
```

```python
my_tuple = ('apple','mango','banana','cherry', 'orange','pineapple')

for x in my_tuple:
    print(x)
```

```python
my_list = []

my_string = str(input("Enter a String: "))
my_list.append(my_string)

for x in my_list:
    print(x)
```

```python


my_string = str(input("Enter a String: "))

for x in my_string:
    print(x)
```

#### 1. Break Statement

```python
my_list = ['apple','mango','banana','cherry', 'orange','pineapple']

for x in my_list:
    print(x)
    if x == "cherry":
        break
```

```python
my_list = ['apple','mango','banana','cherry', 'orange','pineapple']

for x in my_list:
    print(x)
    if x == "mango":
        break
```

```python
my_list = ['apple','mango','banana','cherry', 'orange','pineapple']

for x in my_list:
    print(x)
    if x == "mango":
        break
    
```

#### 2. Continue Statement

```python
my_list = ['apple','mango','banana','cherry', 'orange','pineapple']

for i in my_list:
    if i == "cherry":
        continue
    print(i)
```

### 3. Pass Statement

```python
my_list = ['apple','mango','banana','cherry', 'orange','pineapple']

for i in my_list:
    if i == "cherry":
        pass
    print(i)
```

### 4. Range Function

```python
for x in range(10):
    print(x)
```

```python
for x in range(2, 10):
    print(x)
```

```python
for x in range(3, 30, 3):
    print(x)
```

```python
for x in range(6):
    print(x)
else:
    print("Finished")
```

```python
my_list1 = ['apple','mango','banana','cherry', 'orange','pineapple']
my_list2 = ['red','yellow','yellow','blue','green','gray']


for i in my_list1:
    for j in my_list2:
        print(i, j)
```

```python
for i in range(1, 6):
    for j in range(1, i + 1):
        print(i,end = " ")
    print()
```

```python
for i in range(5, 0,-1):
    for j in range(5,i -1,-1):
        print(j, end=" ")
    print()
```

```python
my_dict = {'name':'Python'}

for i, j in my_dict.items():
    print(i,":", j)
```

```python
my_dict = {
    "Maharashtra": "Akola",
    "Gujrat": "Ahemedabad",
    'Rajstan': "Jaipur",
    'Up':'Ayodhya'
}

for i, j in my_dict.items():
    print(i,"=",j)
```

# 2. While Loop

```python
i = 1

while i < 6:
    print(i)
    i = i + 1
```

```python
i = 1

while i < 10:
    print(i)
    i += 1 #Assignment operator using
```

```python
i = 1
while i < 100:
    print(i)
    if (i == 10):
        break
    i = i + 1
```

```python
i = 1
while i < 100:
    print(i)
    if (i == 10):
        continue
    i = i + 1
```

```python
i = 1
while i < 100:
    print(i)
    i = i + 1
    if (i == 10):
        break
#     i = i + 1
```
