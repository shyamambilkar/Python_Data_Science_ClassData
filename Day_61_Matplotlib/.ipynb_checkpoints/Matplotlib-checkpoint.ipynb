### Marker:
    1. 'o' - circle
    2. '*' - star
    3. '.' - point
    4. ':' - pixel
    5. 'x' - X
    6. '+'- Plus
    7. 's' -square
    8. 'D' - Diamond ['d']
    9. 'p' - Pentagon
    10. 'H' -Hexagon['h']
    11. Tri , down,up, left, right[1,2,3,4]
    

### Line References:
    1. '-' - [solid line]
    2. ':' - [Dotted Line] 
    3. '_' - [Dashed Line]
    4. '-.' -[Dashed / Dotted line]

### Color References:
    1. 'r' - [Red]
    2. 'g' - [Green]
    3. 'b' - [Blue]
    4. 'c' - [ Cyan]
    5. 'm' - [megenta]
    6. 'y' - [Yello]
    7. 'k' - [Black]
    8. 'w' - [White]
